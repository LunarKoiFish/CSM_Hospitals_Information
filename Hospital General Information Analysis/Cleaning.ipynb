```python
import pandas as pd
import numpy as np

```

```python
hospital_df = pd.read_csv("Resources/Hospital_General_Information.csv")
hospital_df.head()
```

```python
hospital_df.info()
hospital_df.isnull().sum() 
```

```python
hospital_df = hospital_df.drop(columns=['Hospital overall rating footnote', 
                                        'MORT Group Footnote', 
                                        'Safety Group Footnote', 
                                        'READM Group Footnote', 
                                        'Pt Exp Group Footnote', 
                                        'TE Group Footnote',
                                        'MORT Group Measure Count',
                                        'Safety Group Measure Count',
                                        'READM Group Measure Count',
                                        'Pt Exp Group Measure Count',
                                        'TE Group Measure Count'
                                        ])
```

```python
hospital_df.head()
```

```python
hospital_df.isnull().sum()
```

```python
hospital_df['Meets criteria for birthing friendly designation']
```

```python
hospital_df['Meets criteria for birthing friendly designation'] = hospital_df['Meets criteria for birthing friendly designation'].fillna('N')
```

```python
hospital_df['Meets criteria for birthing friendly designation']
```

```python
hospital_df.isnull().sum()
```

```python
hospital_df.columns = hospital_df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
hospital_df.columns
```

```python
hospital_df.rename(columns={
    'meets_criteria_for_birthing_friendly_designation':'birthing_friendly'}, inplace=True)
```

```python
hospital_df['birthing_friendly'] = hospital_df['birthing_friendly'].map({'Y': True, 'N': False})
```

```python
hospital_df['emergency_services'] = hospital_df['emergency_services'].map({'Yes': True, 'No': False})
```

```python
num_columns = [
    'hospital_overall_rating',
    'count_of_facility_mort_measures',
    'count_of_mort_measures_better',
    'count_of_mort_measures_no_different',
    'count_of_mort_measures_worse',
    'count_of_facility_safety_measures',
    'count_of_safety_measures_better',
    'count_of_safety_measures_no_different',
    'count_of_safety_measures_worse',
    'count_of_facility_readm_measures',
    'count_of_readm_measures_better',
    'count_of_readm_measures_no_different',
    'count_of_readm_measures_worse',
    'count_of_facility_pt_exp_measures',
    'count_of_facility_te_measures'
]

for col in num_columns:
    hospital_df[col] = pd.to_numeric(hospital_df[col], errors='coerce')
    hospital_df[col] = hospital_df[col].replace([np.inf, -np.inf], pd.NA)
    hospital_df[col] = hospital_df[col].astype('Int64')
```
